```python
#CASE STUDY
# SAMPLE DATA: employee_data = [ {"id": "E001", "name": "Alice Johnson", "department": "Engineering", "salary": 75000}, {"id": "E002", "name": "Bob Smith", "department": "Marketing", "salary": 65000}, {"id": "E003", "name": "Charlie Brown", "department": "Engineering", "salary": 80000} ]

# department_data = [ {"name": "Engineering", "budget": 500000, "employees": ["E001", "E003"]}, {"name": "Marketing", "budget": 300000, "employees": ["E002"]}, {"name": "HR", "budget": 200000, "employees": []} ]

#Problem1

class employee:
    def __init__(self, emp_id, name, department, salary):
        self.info={
            "id":emp_id,
            "name":name,
            "department":department,
            "salary":salary,
        }

    #initializing project list and work history
        self.project=[]
        self.work_history=[]  

    def give_raise(self,amount):
        old_salary=self.info["salary"]
        self.info["salary"] += amount
        self.work_history.append
        (
            f"Salary incresed from {old_salary}to {self.info['salary']}"
        ) 
    def assign_project(self,project_name):
        self.project.append(project_name)
        self.work_history.append(
            f"Assigned project to employee is:{project_name}"
        )
    def get_info(self):
        data=self.info.copy()
        data["project"]=self.project
        data["work_history"]=self.work_history
        return data       

    emp1=employee("E001","Alice Johnson","Engineering",75000)   
    emp2=employee("E002","Bob Smith","Marketing",65000)
    emp3=employee("E003","Charlie Brown","Engineering", 80000)
    
    emp1.assign_project("AIML based project")
    emp2.assign_project("Encryptio based project")
    emp3.assign_project("abc project")

    emp1.give_raise(85000)
    emp2.give_raise(70000)
    emp3.give_raise(85000)

    print(emp1.get_info())
    print(emp2.get_info())
    print(emp3.get_info())
```

```python
#problem 2
class Department:
    def __init__(self, name, budget):
        self.info = {
            "name": name,
            "budget": budget
        }
        self.employees = []

    def add_employee(self, emp_id):
        if emp_id not in self.employees:
            self.employees.append(emp_id)
            print(f"Employee {emp_id} added to {self.info['name']} department.")
        else:
            print(f"Employee {emp_id} already exists in {self.info['name']}.")

    def remove_employee(self, emp_id):
        if emp_id in self.employees:
            self.employees.remove(emp_id)
            print(f"Employee {emp_id} removed from {self.info['name']} department.")
        else:
            print(f"Employee {emp_id} not found in {self.info['name']} department.")

    def get_employee_count(self):
        return len(self.employees)

    def get_info(self):
        data = self.info.copy()
        data["employees"] = self.employees
        return data
    

eng = Department("Engineering", 500000)
mkt = Department("Marketing", 300000)

eng.add_employee("E001")
eng.add_employee("E003")
mkt.add_employee("E002")

eng.add_employee("E001")

mkt.remove_employee("E002")

print("Engineering count:", eng.get_employee_count())
print("Marketing count:", mkt.get_employee_count())

print(eng.get_info())
print(mkt.get_info())

```
