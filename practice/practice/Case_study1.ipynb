```python
#CASE STUDY
# SAMPLE DATA: employee_data = [ {"id": "E001", "name": "Alice Johnson", "department": "Engineering", "salary": 75000}, {"id": "E002", "name": "Bob Smith", "department": "Marketing", "salary": 65000}, {"id": "E003", "name": "Charlie Brown", "department": "Engineering", "salary": 80000} ]

# department_data = [ {"name": "Engineering", "budget": 500000, "employees": ["E001", "E003"]}, {"name": "Marketing", "budget": 300000, "employees": ["E002"]}, {"name": "HR", "budget": 200000, "employees": []} ]

#Problem1

class employee:
    def __init__(self, emp_id, name, department, salary):
        self.info = {
            "id": emp_id,
            "name": name,
            "department": department,
            "salary": salary,
        }

        # initializing project list and work history
        self.project = []
        self.work_history = []

    def give_raise(self, amount):
        old_salary = self.info["salary"]
        self.info["salary"] += amount
        self.work_history.append(
            f"Salary increased from {old_salary} to {self.info['salary']}"
        )

    def assign_project(self, project_name):
        self.project.append(project_name)
        self.work_history.append(
            f"Assigned project to employee: {project_name}"
        )

    def get_info(self):
        data = self.info.copy()
        data["project"] = self.project
        data["work_history"] = self.work_history
        return data



emp1 = employee("E001", "Alice Johnson", "Engineering", 75000)
emp2 = employee("E002", "Bob Smith", "Marketing", 65000)
emp3 = employee("E003", "Charlie Brown", "Engineering", 80000)

# Assign projects
emp1.assign_project("Smart IoT Device")
emp2.assign_project("Secure Cloud Platform")
emp3.assign_project("AI-Based Chatbot")

# Give raises
emp1.give_raise(8500)
emp2.give_raise(7000)
emp3.give_raise(8500)

# Print employee info
print(emp1.get_info())
print(emp2.get_info())
print(emp3.get_info())
```

```python

#problem 2
class Department:
    def __init__(self, name, budget):
        self.info = {
            "name": name,
            "budget": budget
        }
        self.employees = []

    def add_employee(self, emp_id):
        if emp_id not in self.employees:
            self.employees.append(emp_id)
            print(f"Employee {emp_id} added to {self.info['name']} department.")
        else:
            print(f"Employee {emp_id} already exists in {self.info['name']}.")

    def remove_employee(self, emp_id):
        if emp_id in self.employees:
            self.employees.remove(emp_id)
            print(f"Employee {emp_id} removed from {self.info['name']} department.")
        else:
            print(f"Employee {emp_id} not found in {self.info['name']} department.")

    def get_employee_count(self):
        return len(self.employees)

    def get_info(self):
        data = self.info.copy()
        data["employees"] = self.employees
        return data
    

eng = Department("Data management", 500000)
mkt = Department("Marketing", 300000)

eng.add_employee("E001")
eng.add_employee("E003")
mkt.add_employee("E002")

eng.add_employee("E001")

mkt.remove_employee("E002")

print("Data Management count:", eng.get_employee_count())
print("Marketing count:", mkt.get_employee_count())

print(eng.get_info())
print(mkt.get_info())

```

```python
# Problem 3: Company Class

class Company:
    def __init__(self, name):
        self.name = name
        self.employees = {}
        self.departments = {}

    def add_employee(self, employee):
        emp_id = employee.info["id"]
        self.employees[emp_id] = employee
        print(f"Employee {employee.info['name']} added to company {self.name}.")

    def add_department(self, department):
        dept_name = department.info["name"]
        self.departments[dept_name] = department
        print(f"Department {dept_name} added to company {self.name}.")

    def transfer_employee(self, emp_id, new_dept):
        if emp_id not in self.employees:
            print(f"Employee {emp_id} not found in company.")
            return

        employee = self.employees[emp_id]
        old_dept = employee.info["department"]

        if new_dept not in self.departments:
            print(f"Department {new_dept} does not exist.")
            return

        if old_dept in self.departments:
            self.departments[old_dept].remove_employee(emp_id)

        self.departments[new_dept].add_employee(emp_id)
        employee.info["department"] = new_dept

        print(f"Employee {emp_id} transferred from {old_dept} to {new_dept}.")

    def get_total_salary_cost(self):
        total = sum(emp.info["salary"] for emp in self.employees.values())
        return total

    def get_department_summary(self):
        summary = {}
        for dept_name, dept_obj in self.departments.items():
            summary[dept_name] = {
                "budget": dept_obj.info["budget"],
                "employee_count": dept_obj.get_employee_count(),
                "employees": dept_obj.employees
            }
        return summary
    
    

# Create company
comp = Company("Fresh Gravity")

# Create Departments
fia=Department("Finance", 500000)
it = Department("IT department", 300000)
hr = Department("HR", 200000)

comp.add_department(fia)
comp.add_department(it)
comp.add_department(hr)

# Create Employees
e1 = employee("E001", "Alice Johnson", " Data Management", 75000)
e2 = employee("E002", "Bob Smith", "Marketing", 65000)
e3 = employee("E003", "Charlie Brown", "Data Management", 80000)

# Add Employees to Company
comp.add_employee(e1)
comp.add_employee(e2)
comp.add_employee(e3)

# Add Employees to their Departments
fia.add_employee("E001")
it.add_employee("E003")
hr.add_employee("E002")

# Transfer an employee
comp.transfer_employee("E002", "HR")

# Salary raise
e1.give_raise(5000)
e3.give_raise(4000)

# Summary
print("\n Total Salary Cost:", comp.get_total_salary_cost())
print("\n Department Summary:", comp.get_department_summary())

        



```
