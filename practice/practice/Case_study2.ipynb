```python
import pandas as pd

# Load CSV (Full Load)
full_df = pd.read_csv("sales_full.csv")
print("Full Load Data (CSV):")
display(full_df)
```

```python
# Load JSON (Incremental Load)
inc1_df = pd.read_json("_sales_new.json")
inc2_df = pd.read_json("_sales_duplicate.json")

# Display raw incremental files
print(" First Incremental Load (_sales_new.json):")
display(inc1_df)

print(" Second Incremental Load (_sales_duplicate.json, with duplicates):")
display(inc2_df)
```

```python
# Merge all datasets (Full Load + both Incremental Loads)
merged_df = pd.concat([full_df, inc1_df, inc2_df], ignore_index=True)

print("Merged Dataset (Full + Incremental Batches):")
display(merged_df)

print("Total rows before removing duplicates:", len(merged_df))

```

```python
# Remove duplicates based on transaction_id
clean_df = merged_df.drop_duplicates(subset="transaction_id")
print("Cleaned Data (No Duplicate Transactions):")
display(clean_df)
print("Total rows after removing duplicates:", len(clean_df))

```

```python

# Calculate total amount 
clean_df["total_amount"] = clean_df["quantity"] * clean_df["price"]
display(clean_df)

```

```python
# Customer-wise aggregation (total spent per customer)
customer_summary = clean_df.groupby("customer_id")["total_amount"].sum().reset_index()
print(" Customer-wise Total Amount Summary:")
display(customer_summary)

```

```python
# Summary statistics of numeric columns
print("Summary Statistics of the Cleaned Data:")
display(clean_df.describe())
```
